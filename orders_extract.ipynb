#Extract

The file our client is able to download is an Excel file with merged cells, and every order is in a different table with all GLN, address, qty, etc info. It's easy to read in Excel but it's not so easy to read to create our dataframe so, we'll need to transform the file into a dataframe we can use here

```python
orders = pd.read_csv('Order_Example.csv')
orders = orders.T.reset_index(drop=True).T
```

I'm now ready to rename my columns and get rid of rows that I don't need

```python
orders.columns = ['product_category', 'customer', 'po', 'order_date', 'ean', 'buyer_code', 'deliver_to', 'delivery_date', 'item', 'cases', 'qty', 'unit_price', 'order_value']
```

```python
orders = orders.dropna()
```

```python
orders['deliver_to'].unique()
```

```python
orders = orders[orders['product_category'] != 'Seller Code']
orders = orders[orders['ean'] != 'Barcode']
orders = orders[orders['buyer_code'] != 'Buyer Code']
orders = orders[orders['item'] != 'Description']
orders = orders[orders['cases'] != 'Packing']
orders = orders[orders['cases'] != 'Total Qty:']
orders = orders[orders['qty'] != 'Quantity']
orders = orders[orders['unit_price'] != 'Unit Price']
orders = orders[orders['unit_price'] != 'Total Value:']
orders = orders[orders['order_value'] != 'Value']
```
